## 2.2 Importação do Shapefile dos transectos LiDAR

```python
#Definindo o shapefile dos transectos LiDAR sovrevoados no Bioma Amazônia 
voados = "C:/Users/Luciane/Desktop/ser347/Transectos/Voados/transectos_voados_xy.shp"

#Biblioteca geopandas para ler o shapefile de transectos sobrevoados do no Bioma Amazônia
transectos_voados = gpd.read_file(voados)
```

## 3. Interseção dos transectos LiDAR com o estado do Acre

<img src=".\Figuras\metodologia_intersect.png" width = "60%">

```python
#Abrir a tabela de atributos do shapefile de estados do Bioma Amazônia
#Ajuda a verificar como está escrito o estado que será selecionado. 
#Nesse caso iremos selecionar o estado do Acre, que está escrito com letras maíuculas e na coluna NOME.
display(estados)
```

```python
#Seleção do estado do ACRE na coluna NOME.
acre=estados[estados["NOME"]=="ACRE"]
display(acre)
```

```python
#Plotar o shapefile do estado do ACRE

acre.plot(alpha=1, figsize=(12,9),)
plt.title("ACRE", fontsize=24);
plt.grid(b=True, color='gray', linestyle='--', linewidth=0.5);
plt.show()
```

```python
#Interseção dos transectos LiDAR sobrevoados no estado do Acre.
selecionados = transectos_voados[transectos_voados.geometry.intersects(acre.ix[1].geometry)]
```

```python
#Verificação do número de transectos selecionados
#São 19 transectos com 5 colunas na tabela de atributos.
selecionados.shape
```

```python
#Abrir a tabela de atributos do shapefile de transectos sobrevoados no estado do Acre.
display(selecionados)
```

# **4. Gerando a lista de arquivos que serão processados**

```python
import os

#Diretório com os arquivos LiDAR (originais)
dir = "C:\\ser347\\data\\laz"

#Criando o arquivo file_list com os arquivos no diretório definido.
file_list_laz = os.path.join(dir,"file_list_all.txt")

f = open(file_list_laz, 'w')

for index, row in selecionados.iterrows():
    laz_name="{}\\NP_{}.laz\n".format(dir,row['identify'][-4:])
    f.write(laz_name)
f.close()
```
